```python
#numpy library
import numpy as np
import time

lst = list(range(2000000))
arr = np.array(lst)

start = time.time()
lst2 = [x * 2 for x in lst]
print("List: ", time.time() - start)

start = time.time()
arr2 = arr * 2
print("Array: ", time.time() - start)
```

```python
#1d and 2d arrays
arr = np.array([1, 2, 3, 4, 5])
arr2d = np.array([[1, 2, 3], [4, 5, 6], [6, 7, 8]])

print(arr)
print(arr2d)
print(arr2d[0][1])
```

```python
#functions of numpy: zeros(), ones(), full()
nollar = np.zeros((4, 4))
birlar = np.ones((2, 3))
yettilar = np.full((2, 3), 7)

print(nollar,"\n\n", birlar, "\n\n", yettilar)
```

```python
#arange() and linspace()
l1 = np.arange(0, 31, 2) #(start, stop, step)
l2 = np.linspace(1, 31, 15) #(start, stop, amount)

print(l1)
print(l2)
```

```python
#random library in numpy (rand() vs randint())
np.random.seed(42) #seed() used when we dont want to change the values
a = np.random.rand(3, 3) #float from 0 to 1
b = np.random.randint(0, 10, (3, 3)) #butun sonlar: from 0 to 10

print(a)
print(b)
```

```python
#attributes of array
onlar = np.full((5, 5), 10)
print(onlar)
print(onlar.shape) #number of rows and cols
print(onlar.size) #rows*cols
print(onlar.ndim) #dimeansion
print(onlar.dtype) #data type
```

```python
#Indexing vs Slicing

#Indexing - indekslash
#Slicing - o'zimizga kerak joyini kesib olish

arr = np.array([1, 2, 3, 4])
print(arr[0])
print(arr[-1])
print(arr[1:4])
print(arr[::2]) #step=2
print(arr[2::]) #oxirgi 2tasi
```

```python
arr = np.random.randint(0, 10, (1,5))
print(arr)
print(f"Sum: {np.sum(arr)}")
print(f"Mean: {np.mean(arr)}")
print(f"Standard Division: {np.std(arr)}")
print(f"Minimum Value: {np.min(arr)}")
print(f"Maximum Value: {np.max(arr)}")
print(f"Maximum Index: {np.argmax(arr)}") #argmin()
```
